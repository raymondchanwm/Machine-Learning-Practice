```python
import pandas as pd
train = pd.read_csv("../input/house-prices-advanced-regression-techniques/train.csv")
test = pd.read_csv("../input/house-prices-advanced-regression-techniques/test.csv")
print('train shape: ', train.shape)
print('test:', test.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew
%matplotlib inline
prices = pd.DataFrame({'price':train["SalePrice"], 
                       'log(price+1)':np.log1p(train['SalePrice'])})
print(prices, '\n')
print('price skew: \t', skew(prices['price']))
print('log(price+1) skew: \t', skew(prices['log(price+1)']))
plt.rcParams['figure.figsize'] = (12.0, 6.0)
prices.hist()
```

```python
train.info()
```

```python
train["SalePrice"] = np.log1p(train["SalePrice"])
```

```python
all_data = pd.concat((train.loc[:,"MSSubClass":"SaleCondition"],test.loc[:,"MSSubClass":"SaleCondition"]))
print(all_data.shape)
print(all_data)
```

```python
from scipy.stats import skew
numeric_feats = all_data.dtypes[all_data.dtypes != "object"].index
print("-----Column of non-object type-----")
print(numeric_feats)
skewed_feats = train[numeric_feats].apply(lambda x: skew(x.dropna()))
print("-----Skewness of non-object type column-----")
print(skewed_feats)
skewed_feats = skewed_feats[skewed_feats > 0.75]
print("-----Skewness greater than 0.75-----")
print(skewed_feats)
skewed_feats = skewed_feats.index
all_data[skewed_feats] = np.log1p(all_data[skewed_feats])
all_data[skewed_feats]
```

```python
cc_data = pd.get_dummies(train["LotShape"])
cc_data["LotShape"] = train["LotShape"]
cc_data[:20]
```

```python
all_data = pd.get_dummies(all_data)
```

```python
all_data = all_data.fillna(all_data[:train.shape[0]].mean())
```

```python
X_train = all_data[:train.shape[0]]
X_test = all_data[train.shape[0]:]
y = train.SalePrice
```

```python
from sklearn.model_selection import cross_val_score
def rmse_cv(model):
    """
    Parameters:
        model(obj): Model object
    returns:
        (float) rmse of output and true value
    """
    rmse = np.sqrt(-cross_val_score(model, X_train, y, scoring = "neg_mean_squared_error", cv = 5)) #split = 5
    return rmse
```

```python
from sklearn.linear_model import Ridge

#building ridge model
model_ridge = Ridge()

#set 10 L2 strength
alphas = [0.05, 0.1, 0.5, 1, 5, 10, 15, 30, 50, 75]

#find rmse
cv_ridge = [rmse_cv(Ridge(alpha = alpha)).mean() for alpha in alphas]

#set cv_ridge to series object
cv_ridge = pd.Series(cv_ridge, index = alphas)
print("Ridge RMSE Loss: ")
print(cv_ridge, "\n")

#print average loss
print('Ridge RMSE loss Mean')
print(cv_ridge.mean(), "\n")

#plot graph
plt.figure(figsize = (10,5)) #set plotting size
plt.plot(cv_ridge) #plot cv_ridge
plt.grid()
plt.title("Validation - by regulariztion strength")
plt.xlabel("Alpha")
plt.ylabel("RMSE")
plt.show()
```

```python
from sklearn.linear_model import LassoCV

#using 4 diff LASSO
model_lasso = LassoCV(alphas = [1, 0.1, 0.001, 0.0005]).fit(X_train,y)

print('Lasso regression RMSE loss:')
cv_lasso = rmse_cv(model_lasso)
print(rmse_cv(model_lasso))

print("Average loss:", cv_lasso.mean())
print("Minimun loss:", cv_lasso.min())
print("Best alpha: ", model_lasso.alpha_)
```

```python
import xgboost as xgb

dtrain = xgb.DMatrix(X_train, label = y)

#set learning rate and decision tree's depth
params = {"max_depth":3, "eta":0.1}

#xgboost model and cross validate
cross_val = xgb.cv(params, 
                   dtrain, 
                   num_boost_round = 1000, 
                   early_stopping_rounds = 50)
cross_val
```

```python
plt.figure(figsize = (8,6))
plt.plot(cross_val.loc[30:, ["train-rmse-mean"]], 
         linestyle = '--', 
         label = "Train")
plt.plot(cross_val.loc[30:, "test-rmse-mean"],
         label = 'Validation')
plt.grid()
plt.xlabel("num_boost_rounds")
plt.ylabel("RMSE")
plt.legend()
plt.show()
```

```python
model_xgb = xgb.XGBRegressor(n_estimators = 410, 
                            max_depth = 3, 
                            learning_rate = 0.1)
model_xgb.fit(X_train, y)
print("xgboost RMSE loss:" )
print(rmse_cv(model_xgb).mean())
```

```python
lasso_preds = np.expm1(model_lasso.predict(X_test))
xgb_preds = np.expm1(model_xgb.predict(X_test))
```

```python
preds = lasso_preds * 0.7 + xgb_preds * 0.3
```

```python
solution = pd.DataFrame({"id":test.Id, 
                         "SalePrice":preds})
solution.to_csv("emsemble_sol.csv", index = False)
```
